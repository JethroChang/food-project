```python
#抓每個城市的內容頁網址與店家名稱
import requests
from bs4 import BeautifulSoup
import concurrent.futures
import time


url = "https://www.mobile01.com/waypointtopiclist.php?f=191&p=1"
res = requests.get(url)
soup = BeautifulSoup(res.text, 'lxml')
info = soup.select(".pagination > a")
totalpage = int(info[len(info)-1].text)
#抓取總頁數

for i in range(1, totalpage+1):
    url = "https://www.mobile01.com/waypointtopiclist.php?f=191&p="+str(i)
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'lxml')
    info = soup.select(".subject-text > a")
    hrefli = []
    for hr in info:
        hrefli.append("https://www.mobile01.com/"+hr['href'])
    nameli = []
    for name in info:    
        nameli.append(name.text)    
    info2 = soup.select(".authur > a > p")
    timeli = []
    for i in range(0,(len(info2))):
        if i%3 == 0:
            timeli.append(info2[i].text)
        else:
            pass
    li = []
    for i in range(0,(len(hrefli))):
        li.append(hrefli[i])
        li.append(nameli[i])
        li.append(timeli[i])
        
                       
    with open('mobile01-Taoyuan.txt', 'a', encoding = 'UTF-8') as f:
        for tx in li:  
            f.write(tx)  
            f.write('\n')
    time.sleep(3)
```
